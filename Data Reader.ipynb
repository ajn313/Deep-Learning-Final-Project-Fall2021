```python
#code to parse and compile the data

first = 0
base = []
for i in range(len(urls)):
    curr_request = req.Request(url=urls[i], headers=headers) #request and read the URL
    curr_ratings = req.Request(url=ratings[i], headers=headers)
    content = req.urlopen(curr_request)
    rate = req.urlopen(curr_ratings)
    parse = BeautifulSoup(content, 'html.parser') #parse with Beautiful Soup
    parser = BeautifulSoup(rate, 'html.parser')
    name = parser.find('h1', attrs={'id': 'title'})
    scores = parser.find_all('td', attrs={'class': 'score'})
    numscore = []
    for i in range (len(scores)):                  #create array with score for each category
        scores[i]=scores[i].get_text().strip()
        if (scores[i]!='N/A' and scores[i]!='nan'):
            numscore.append(float(scores[i]))
        else:
            continue
    rating = np.mean(numscore)  #take the average of the scores and round
    rating = round(rating)
    text1 = parse.find_all('span', attrs={'class': 'key'})   #find the key and value for each specification
    text2 = parse.find_all('span', attrs={'class': 'value'})
    text1clean = []
    text2clean = []
    text1clean.append("Name")
    text1clean.append("Rating")
    text2clean.append(name.get_text().strip())
    text2clean.append(rating)

    for i in text1:
        text1clean.append(i.get_text().strip())
    for i in text2:
        text2clean.append(i.get_text().strip())

    if first == 0:                                           #write it all to a CSV file
        with open('cars.csv','a',newline = "") as csv_file:
          writer = csv.writer(csv_file,delimiter=',',quotechar='"', quoting = csv.QUOTE_ALL)
          writer.writerow(text1clean)
        first = 1
        base = text1clean
    curr = []
    for i in base:
        try:
            ind = text1clean.index(i)
            add = text2clean[ind]
        except ValueError:
            add = "N/A"
        curr.append(add)
    with open('cars.csv','a',newline = "") as csv_file:
          writer = csv.writer(csv_file,delimiter=',',quotechar='"', quoting = csv.QUOTE_ALL)
          writer.writerow(curr)
    
```
